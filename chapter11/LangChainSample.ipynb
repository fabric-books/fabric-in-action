```python
from langchain.vectorstores import FAISS
from langchain.schema import Document
from langchain_huggingface import HuggingFaceEmbeddings

docs = [Document(page_content=t) for t in [
    "LangChain helps developers build LLM-powered applications.",
    "Microsoft Fabric unifies data and AI workflows.",
    "Vector search retrieves semantically similar documents.",
    "Hugging Face hosts open-source language models."
]]

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.from_documents(docs, embedding)
query = "How do I use LangChain?"

for i, (doc, score) in enumerate(db.similarity_search_with_score(query, k=2), 1):
    v = embedding.embed_query(doc.page_content)
    print(f"{i}. {doc.page_content}\n   Score: {score:.4f}\n   Vector: {v[:5]}\n")
```
