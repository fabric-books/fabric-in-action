```python
import pandas as pd
import folium

# Prepare and rename DataFrame columns
df = pd.DataFrame(df100).rename(columns={
    0: 'Cities',
    1: 'Latitude',
    2: 'Longitude',
    3: 'Hotels',
    4: 'Rating',
    5: 'Reviews'
})
df['Rating'] = df['Rating'].astype('int8')

# Initialize the map centered on the USA
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Add circle markers based on hotel ratings
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5 * row['Rating'],
        popup=f"{row['Hotels']}, Rating: {row['Rating']}",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(m)

m
```
